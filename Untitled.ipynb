```python
import pandas as pd
yeast = pd.read_csv(r"C:\Users\alexh\Desktop\yeast_train.csv")
yeast
```

```python
import numpy as np

a = np.array(yeast.F7)
b = np.array(yeast.F8)
new = np.column_stack((a,b))
new
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(new, yeast.L, test_size=0.35, random_state=42)

SXtrain, Xvalid, Sytrain, vtest = train_test_split(Xtrain, ytrain, test_size=0.3,
                                                random_state=42)
```

```python
from sklearn import preprocessing

scaler = preprocessing.StandardScaler().fit(SXtrain)

"""Scale the smaller training set"""
scaled_trnX = scaler.transform(SXtrain) 
"""Scale the validation set"""
scaled_valX = scaler.transform(Xvalid) 
```

```python
from sklearn.svm import SVC

svc1 = SVC(kernel='rbf', class_weight='balanced', C=10, gamma=0.1)
model1 = svc1.fit(scaled_trnX, Sytrain)

"""The fitted model should be validated on the scaled validation set. """
vyfit1 = model1.predict(scaled_valX)

"""Performance measurements"""
from sklearn import metrics 
print('Accuracy:', metrics.accuracy_score(vtest, vyfit1))

from sklearn.metrics import classification_report
print(classification_report(vtest, vyfit1,
                            labels=yeast.L))
```

```python
suitable_gamma = 0.1
suitable_cost = 10
```

```python
scaler = preprocessing.StandardScaler().fit(Xtrain)

scaled_X = scaler.transform(Xtrain)      
scaled_tstX = scaler.transform(Xtest) 
```

```python
svc_final = SVC(kernel='rbf', class_weight='balanced',C=suitable_cost, gamma=suitable_gamma)
model_final = svc_final.fit(scaled_X, ytrain)
yfit_test = model_final.predict(scaled_tstX)

from sklearn import metrics 
print('Accuracy:', metrics.accuracy_score(ytest, yfit_test))

print(classification_report(ytest, yfit_test,
                            labels=yeast.L))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
#use Seaborn plotting defaults
import seaborn as sns; sns.set()

mat = confusion_matrix(ytest, yfit_test)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=yeast.L,
            yticklabels=yeast.L)
plt.xlabel('true label')
plt.ylabel('predicted label');
```
